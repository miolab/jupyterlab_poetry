# Spectral Clustering sample (Non Hierarchical Clustering)
- Cluster creation from __data density__.
- Therefore, it works even in non-linear situations.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import cluster
from sklearn import preprocessing
from sklearn.datasets import make_moons


X, z = make_moons(
    n_samples=200,
    noise=0.05,
    random_state=0
)

sc = preprocessing.StandardScaler()
X_norm = sc.fit_transform(X)
display(X_norm)
```

```python
x = X_norm[:, 0]
y = X_norm[:, 1]

plt.figure(figsize=(8, 3))
plt.scatter(x, y, c=z)
plt.show()
```

```python
km = cluster.KMeans(
    # Specified 2 clusters to divide into upper and lower.
    n_clusters=2,
    n_init=10
)
z_km = km.fit(X_norm)

plt.figure(figsize=(8, 3))
plt.scatter(
    x, 
    y, 
    c=z_km.labels_
)
plt.scatter(
    z_km.cluster_centers_[:, 0],
    z_km.cluster_centers_[:, 1],
    s=250,
    marker='*',
    c='red'
)    
plt.show()
```
