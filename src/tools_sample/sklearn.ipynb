```python
import sklearn.datasets
import sklearn.linear_model
import sklearn.model_selection
import pandas as pd
import matplotlib.pyplot as plt

diabetes = sklearn.datasets.load_diabetes()

df = pd.DataFrame(
    diabetes.data,
    columns=diabetes.feature_names
)
df
```

```python
# sample: Numerical data on disease progression after 1 year.
diabetes.target
```

```python
""" Cross-validation sample
"""
X = diabetes.data
y = diabetes.target

X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(
    X, y, test_size=0.2
)

# Objects for machine learning with linear regression algorithms.
lr = sklearn.linear_model.LinearRegression()

# Pass training data to `fit` method and execute machine learning.
lr.fit(X_train, y_train)

# Show accuracy of machine learning results.
lr.score(X_test, y_test)
```

```python
# Machine Learning Prediction.
predicted = lr.predict(X)

# Visualize the difference between actual result and Machine Learning Prediction.
figure, axis = plt.subplots()
axis.scatter(
    y,
    predicted,
    c='b',
    edgecolors=(0, 0, 0)
)
axis.plot(
    [y.min(), y.max()],
    [y.min(), y.max()],
    "g--",
    lw=3
)
axis.set_xlabel("Measured")
axis.set_ylabel("Predicted")
plt.show()
```
