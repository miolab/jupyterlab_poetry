```python
"""Matplotlib samples
"""
import numpy as np
import matplotlib.pyplot as plt

# Draw a line chart with elements on Y-axis and indices on X-axis.
plt.plot(
    [1, 2, 3, 4, 5]
)
plt.show()
```

```python
plt.plot(
    # X-axis to Y-axis
    [1, 2, 3, 4, 5],
    [1, 4, 9, 16, 25]
)
plt.show()
```

```python
# Draw a graph with round dots.
plt.plot(
    [1, 2, 3, 4, 5],
    [1, 4, 9, 16, 25],
    'o'
)
plt.show()
```

```python
plt.plot(
    [1, 2, 3, 4, 5],
    [1, 4, 9, 16, 25],
    'o'
)
# Specify the X-axis and Y-axis ranges.
plt.axis(
    # X-min, X-max, Y-min, Y-max 
    [0, 3.5, 0, 15]
)
plt.show()
```

```python
arr = np.arange(0, 10, 0.5)

# Draw multiple graphs in a single figure.
plt.plot(
    arr,
    arr,
    # green and dashed line
    'g--',
    label='y = x'
)
plt.plot(
    arr,
    arr * 2,
    # blue and upward triangle dots
    'b^',
    label='y = x * 2'
)
plt.plot(
    arr,
    arr ** 2,
    # red and square dots
    'rs',
    label='y = x ** 2'
)

# Displays a legend with the each specified label name.
plt.legend()

plt.show()
```

```python
# Scatterplot sample diagram.
x_axis = np.random.rand(100)
y_axis = np.random.rand(100)

plt.scatter(x_axis, y_axis)
plt.show()
```

```python
# Colormap sample.
x_axis_2 = np.random.rand(100)
y_axis_2 = np.random.rand(100)

plt.scatter(
    x_axis_2,
    y_axis_2,
    # set size and color
    # (e.g) s=100, c='r'
    s=(np.random.rand(100) * 130),
    c=np.random.rand(100),
)
plt.show()
```

```python
# Bar graph sample.
values = [
    1, 2, 3, 4, 5
]
x_axis = np.arange(
    len(values)
)

plt.bar(x_axis, values)

# optional: set X-axis each item names
plt.xticks(
    x_axis,
    ['a', 'b', 'c', 'd', 'e']
)

plt.show()
```
