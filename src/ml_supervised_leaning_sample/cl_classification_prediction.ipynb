```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
from sklearn.ensemble import RandomForestClassifier
from mlxtend.plotting import plot_decision_regions


def load_data():
    """Load breast cancer dataset.
    - ref: https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_breast_cancer.html
    """
    return load_breast_cancer()


def construct_logistic_regression(input_x, input_y):
    """Construct Logistic Regression classifier model.
    """
    return LogisticRegression(random_state=0).fit(
        input_x,
        input_y
    )


def construct_linear_svms(input_x, input_y):
    """Construct Linear SVM classifier model.
    """
    return LinearSVC(random_state=0).fit(
        input_x,
        input_y
    )


def construct_kernel_svms(input_x, input_y):
    """Construct Kernel SVM classifier model.
    """
    return SVC(
        kernel='rbf',
        random_state=0
    ).fit(
        input_x,
        input_y
    )


def construct_k_nearest_neighbors_classifier(input_x, input_y, neighbors):
    """Construct K-nearest neighbors classifier model.
    """
    return KNeighborsClassifier(
        n_neighbors=neighbors,
        p=2
    ).fit(
        input_x,
        input_y
    )


def construct_decision_tree_classifier(
    tree_classifier,
    input_x,
    input_y,
    depth: int = 3
):
    """Construct Decision Tree classifier model.
    - DecisionTreeClassifier
    - RandomForestClassifier
    """
    return tree_classifier(
        max_depth=depth,
        random_state=0
    ).fit(
        input_x,
        input_y
    )


def plot_scatter(
    target,
    x_direcsion_col: str,
    y_direction_col: str,
    y_elements,
    title: str = ''
) -> None:
    """Visulalize data scatter."""
    plt.scatter(
        target[x_direcsion_col],
        target[y_direction_col],
        c=y_elements
    )
    plt.title(title)
    plt.xlabel(x_direcsion_col)
    plt.ylabel(y_direction_col)
    plt.show()


def draw_decision_regions(input_x, input_y, regression_model) -> None:
    """Plot decision region."""
    plot_decision_regions(
        np.array(input_x),
        np.array(input_y),
        clf=regression_model
    )
    plt.show()
```

```python
# Prepare dataset.
df = pd.DataFrame(
    load_data().data,
    columns=load_data().feature_names
)
df['y'] = load_data().target

display(df)
```

```python
target_df = df[
    ['mean radius', 'mean texture', 'y']
]
display(target_df.describe())
```

Each count are same number, so there seems to be NO __missing values__.

```python
display(target_df.corr())
```

__Correlations__ between the variables used seem to be just right.

```python
# Check the number of categories for the objective variable.
print(target_df['y'].unique())
```

```python
print(len(
    df.loc[target_df['y'] == 0]
))
print(len(
    df.loc[target_df['y'] == 1]
))
```

The category ratios are well balanced.

```python
# Visulalize data scatter.
plot_scatter(
    target_df,
    'mean radius',
    'mean texture',
    target_df['y']
)
```

```python
# Split the data into Explanatory and Objective variables.
X = target_df[
    ['mean radius', 'mean texture']
]
y = target_df[
    'y'
]

display(X.head())
display(y.head())
```

```python
# Split dataset into training and test data.
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=0
)

print(f'X_train length: {len(X_train)}')
display(X_train.head())
print(f'\nX_test length: {len(X_test)}')
display(X_test.head())
```

```python
# Align data scales.
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print(X_train_scaled[:5])
print(X_test_scaled[:5])
```

```python
# Construct Logistic Regression classifier model.
logistic_regression = construct_logistic_regression(
    X_train_scaled,
    y_train
)
```

```python
# Show predicted values.
y_train_predict = logistic_regression.predict(X_train_scaled)
y_test_predict = logistic_regression.predict(X_test_scaled)

print(f'y_train_predict: {len(y_train_predict)}\n{y_train_predict}\n')
print(f'y_test_predict: {len(y_test_predict)}\n{y_test_predict}')
```

```python
# Plot predict train results.
plot_scatter(
    X_train,
    'mean radius',
    'mean texture',
    y_train_predict,
    'Predict Train'
)
```

```python
# Plot predict train results.
plot_scatter(
    X_test,
    'mean radius',
    'mean texture',
    y_test_predict,
    'Predict Test'
)
```

```python
# Plot the Logistic Regression Decision region.
draw_decision_regions(
    X_train_scaled,
    y_train,
    logistic_regression
)
```

```python
# Construct Linear SVMs classification model.
linear_svm = construct_linear_svms(
    X_train_scaled,
    y_train
)

draw_decision_regions(
    X_train_scaled,
    y_train,
    linear_svm
)
```

The results are often similar to __Logistic regression__ and __Linear SVM__.

- __Logistic regression__ is a __probability theory based__ algorithm. 
- __Linear SVM__ is a that focusing on margin maximization algorithm.
  - Draw a decision region that maximizes the distance to the __Support vector__ (= the data point closest to the decision boundary).

Linear SVM is less sensitive to data outliers than Logistic regression.

```python
# Construct Kernel SVMs classification model.
kernel_svm = construct_kernel_svms(
    X_train_scaled,
    y_train
)

draw_decision_regions(
    X_train_scaled,
    y_train,
    kernel_svm
)
```

Use __Kernel method__ with SVM.
- Kernel method is a method for converting data that cannot be linearly separated into a state that can be linearly separated.

```python
# Construct K-nearest neighbors algorithm model.
k_n_classifier = construct_k_nearest_neighbors_classifier(
    X_train_scaled,
    y_train,
    5
)

draw_decision_regions(
    X_train_scaled,
    y_train,
    k_n_classifier
)
```

K-nearest neighbors algorithm

- It memorizes the teacher data and derives the classification prediction results for the data to be predicted by majority voting on which of the K training data in the vicinity is classified.
- Therefore, learning is fast, but prediction takes time. (Computational cost is high.)
- The advantage is that the models are easy to interpret and can be applied rather flexibly to any data set.

```python
# Construct Decisino Tree classification model.
decision_tree_classifier = construct_decision_tree_classifier(
    DecisionTreeClassifier,
    X_train_scaled,
    y_train
)

draw_decision_regions(
    X_train_scaled,
    y_train,
    decision_tree_classifier
)
```

```python
# Visualize decision tree.
plt.figure(figsize=(20, 8))
plot_tree(
    decision_tree_classifier,
    feature_names=['mean radius', 'mean texture'],
    filled=True,
    proportion=True,
    fontsize=12
)
plt.show()
```

- Can be presented as a basis for forecasting results.
- There is a disadvantage in terms of accuracy, as it is easy to fall into over-learning.

```python
# Construct Random Forest classification model.
random_forest_classifier = construct_decision_tree_classifier(
    RandomForestClassifier,
    X_train_scaled,
    y_train
)

draw_decision_regions(
    X_train_scaled,
    y_train,
    random_forest_classifier
)
```
