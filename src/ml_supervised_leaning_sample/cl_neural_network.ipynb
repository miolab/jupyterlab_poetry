# Classification prediction (Neural Network)

- sklearn.neural_network.MLPClassifie https://scikit-learn.org/stable/modules/generated/sklearn.neural_network.MLPClassifier.html

```python
%matplotlib inline
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions
from sklearn.datasets import make_circles
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
```

```python
# Prepare dataset.
X, y = make_circles(
    random_state=48,
    n_samples=200,
    noise=0.1,
    factor=0.3
)

plt.figure(figsize=(6, 4))
plt.scatter(
    X[:, 0],
    X[:, 1],
    c=y
)
plt.show()
```

```python
classifier = MLPClassifier(
    random_state=0
)
model = classifier.fit(X, y)

print(classifier)
```

```python
plt.figure(figsize=(8, 4))
plot_decision_regions(
    X,
    y,
    clf=model
)
plt.show()
```
